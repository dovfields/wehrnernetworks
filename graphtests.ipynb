```python
import networkx as nx
import matplotlib.pyplot as plt
import random as random
import numpy as np
%matplotlib inline
```

```python
#parameters
v = 10 # of vertices
p = 0.3 # prob of connection
```

```python
#Erdos-Renyii graph
G = nx.gnp_random_graph(v,p)
#add random weights btw 0 and 1
for e in G.edges():
    G.edges[e]['weight'] = random.random()
#draw graph
posdef = nx.spring_layout(G)
nx.draw(G,posdef)
elabel = {}
for edge in G.edges():
    label = round(G.edges[edge]['weight'],2)
    xi,yi = posdef[edge[0]]
    xf,yf = posdef[edge[1]]
    elabel[label] = ((xf + xi)/2,(yf + yi)/2)
for label in elabel:
    x,y = elabel[label]
    plt.text(x,y,label)
```
